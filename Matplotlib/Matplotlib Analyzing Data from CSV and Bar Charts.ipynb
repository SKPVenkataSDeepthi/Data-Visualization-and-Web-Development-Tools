```python
import csv ## importing csv is module 
import numpy as np ## importing numpy module 
from collections import Counter
from matplotlib import pyplot as plt ### matplotlib is a lib from which you are importing a module called pyplot 

plt.style.use("fivethirtyeight")

with open('data.csv') as csv_file: ## opening the csv file 
    csv_reader = csv.DictReader(csv_file) ## read the csv file
    
    language_counter = Counter()
    
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
        
language_counter
```

```python
# gives most top responses and plotting of the graph 
languages = []
popularity = []
for item in language_counter.most_common(10): 
    languages.append(item[0])
    popularity.append(item[1])

print(languages)
print(popularity)

plt.bar(languages, popularity)
plt.title("Most popular langauges")
plt.xlabel('Programming Languages')
plt.ylabel('Number of people who use')
plt.show()
```

```python
# as the graph above is clumpsy we solve it using barh
# barh === horizontal method to make it wasy to read , leave aruguments as usuall and change axis labels 
languages = []
popularity = []

for item in language_counter.most_common(15): 
    languages.append(item[0])
    popularity.append(item[1])

print(languages)
print(popularity)

plt.barh(languages, popularity)

plt.title("Most popular langauges")
plt.ylabel('Programming Languages')
plt.xlabel('Number of people who use')
plt.show()
```

```python
# to reverse it we use reverse() --- it automatically reverses in place
languages = []
popularity = []

for item in language_counter.most_common(10): 
    languages.append(item[0])
    popularity.append(item[1])

languages.reverse()
popularity.reverse()

plt.barh(languages, popularity)
plt.title("Most popular langauges")
plt.ylabel('Programming Languages')
plt.xlabel('Number of people who use')
plt.show()
```
