```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

slices = [120, 80, 30, 20] ##these are the values that represent your pie charts called as slices 
labels = ['Sixty', 'Forty', 'Extra1', 'Extra2']
colors = ['blue', 'red', 'yellow', 'green']

plt.pie(slices, labels = labels, wedgeprops = {'edgecolor':'black'})
plt.title("Pie-mazing Data Adventures!")
plt.tight_layout()
plt.show()
```

```python
# if you have large chunks of data like mentioned below go with bar graphs.. else with pie
## seperate on particular slice of the p;ie 
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

slices = [59219, 55466, 47544, 36443,35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0] ### seperate one parlticular slice of the pie. 

plt.pie(slices, labels = labels, explode = explode,shadow = True, wedgeprops = {'edgecolor':'black'})
plt.title("Pie-mazing Data Adventures!")
plt.tight_layout()
plt.show()
```

```python
#to rotate the pie chart 
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

slices = [59219, 55466, 47544, 36443,35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0] 

plt.pie(slices, labels = labels, explode = explode,shadow = True,startangle = 90, wedgeprops = {'edgecolor':'black'})
plt.title("Pie-mazing Data Adventures!")
plt.tight_layout()
plt.show()
```

```python
## display the % each slice takes up 
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

slices = [59219, 55466, 47544, 36443,35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0] ### seperate one parlticular slice of the pie. 

plt.pie(slices, labels = labels, explode = explode,shadow = True,startangle = 90, autopct = '%1.1f%%', wedgeprops = {'edgecolor':'black'})
plt.title("Pie-mazing Data Adventures!")
plt.tight_layout()
plt.show()
```
